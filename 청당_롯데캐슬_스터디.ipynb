```python
import pandas as pd
from google.colab import files
import openpyxl
from openpyxl.styles import PatternFill, Border, Side, Alignment, Font
from openpyxl.utils.dataframe import dataframe_to_rows

# 파일 업로드
uploaded = files.upload()
file_name = next(iter(uploaded))

# 각 시트 읽기
input_df = pd.read_excel(file_name, sheet_name='input', header=None)
donghosu_df = pd.read_excel(file_name, sheet_name='동호수')
bojeongchi_df = pd.read_excel(file_name, sheet_name='보정치', header=1)  # 첫 번째 행을 제외하고 읽기

# 'input' 시트 데이터 처리
# (여기에 'input' 시트 데이터 처리 로직 추가 - 현재는 input_df를 그대로 사용)

# '동호수' 시트 데이터 처리
# (여기에 '동호수' 시트 데이터 처리 로직 추가)

# '보정치' 시트 데이터 처리
bojeongchi_df = bojeongchi_df.applymap(lambda x: f"{x*100}%" if isinstance(x, (int, float)) else x)

# 출력 파일명 지정 (확장자 자동 추가)
output_file_name = input("Enter the name of the output excel file: ")
if not output_file_name.endswith('.xlsx'):
    output_file_name += '.xlsx'

# 엑셀 파일로 저장
with pd.ExcelWriter(output_file_name, engine='openpyxl') as writer:
    input_df.to_excel(writer, index=False, header=False, sheet_name='output')
    donghosu_df.to_excel(writer, index=False, sheet_name='동호수output')
    bojeongchi_df.to_excel(writer, index=False, sheet_name='보정치')

# 엑셀 스타일 조정
wb = openpyxl.load_workbook(output_file_name)

# 공통 스타일 설정
thin_border = Border(left=Side(style='thin'), right=Side(style='thin'), top=Side(style='thin'), bottom=Side(style='thin'))
center_aligned_text = Alignment(horizontal='center')
header_fill = PatternFill(start_color="FFFF00", end_color="FFFF00", fill_type="solid")  # 노란색으로 변경
category_fill = PatternFill(start_color="DDDDDD", end_color="DDDDDD", fill_type="solid")

# 'output' 시트 스타일 설정
ws_output = wb['output']
for row in ws_output.iter_rows():
    for cell in row:
        cell.border = thin_border
        cell.alignment = center_aligned_text
    row[0].fill = header_fill  # A열에 색상 적용

# '보정치' 시트 스타일 설정
ws_bojeongchi = wb['보정치']
for row in ws_bojeongchi.iter_rows(min_row=2, max_col=2, max_row=ws_bojeongchi.max_row):
    for cell in row:
        cell.border = thin_border
        cell.alignment = center_aligned_text
        if cell.column in [2, 3]:  # B열과 C열에 해당하는 셀
            cell.fill = category_fill

# '보정치' 시트 제목 행 색상 채우기
for cell in ws_bojeongchi[1]:  # 첫 번째 행 (제목 행)
    cell.fill = header_fill

# '동호수output' 시트 제목 행 색상 채우기 및 백분율 설정
ws_donghosu_output = wb['동호수output']
for cell in ws_donghosu_output[1]:  # 첫 번째 행 (제목 행)
    cell.fill = category_fill
    if cell.value in ['층지수', '향지수', '구조지수', '평형지수', '지수종합']:
        for row in ws_donghosu_output.iter_rows(min_row=2, max_row=ws_donghosu_output.max_row):
            row[cell.column - 1].number_format = '0.00%'

wb.save(output_file_name)

# 생성된 파일 다운로드
files.download(output_file_name)
```
