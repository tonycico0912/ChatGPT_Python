```python
# seaborn  sns  . seaborn    .
import seaborn as sns

# matplotlib pyplot  plt  .     .
import matplotlib.pyplot as plt

# seaborn  load_dataset   'penguins'  ,  df  .
df = sns.load_dataset('penguins')

#   .
print(df)

# scatter plot . x 'flipper_length_mm'  , y 'body_mass_g'  .
plt.scatter(df['flipper_length_mm'], df['body_mass_g'])

#   .
plt.show()

# 'species'  ,   'body_mass_g'  .     df_group .
df_group = df.groupby('species')['body_mass_g'].mean().reset_index()

# bar plot . x 'species', y 'body_mass_g' .
plt.bar(x=df_group['species'], height=df_group['body_mass_g'])

#   .
plt.show()

# matplotlib   'Malgun Gothic' .     .
plt.rc('font', family='Arial')

#  . 'body_mass_g'  30   .
plt.hist(df['body_mass_g'], bins=30)

# x  'Body Mass' .
plt.xlabel('Body Mass')

# y  'Count' .
plt.ylabel('Count')

#   'distribution of Penguin's weight' .
plt.title('distribution of weight of Penguins')

#   .
plt.show()
```
