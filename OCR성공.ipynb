```python
import requests
import datetime
import json
from google.colab import files

# API 정보
api_url = "https://neowbc3ots.apigw.ntruss.com/custom/v1/23393/955bd967d5bc739d17dc5ece5ccdaf734227cad2782b669b786febbfb5bf3d92/general"
x_ocr_secret = "UHVzQ0ZFamp4ZWdYVUZsQlp4UWhyQUtkUnVIb2FTRXE="  # X-OCR-SECRET

# 파일 업로드
uploaded = files.upload()
for fn in uploaded.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(
      name=fn, length=len(uploaded[fn])))

# API에 전달할 데이터
data = {
    "version": "V2",
    "requestId": "string",
    "timestamp": int(datetime.datetime.now().timestamp()),
    "images": [
        {
            "format": fn.split('.')[-1],  # 확장자 추출
            "name": "medium"
        }
    ]
}

# API에 전달할 헤더
headers = {
    "X-OCR-SECRET": x_ocr_secret
}

# 파일 데이터
file_data = {
    'file': (fn, open(fn, 'rb'))
}

# POST 요청을 보냅니다.
response = requests.post(api_url, headers=headers, data={'message': json.dumps(data)}, files=file_data)

# 요청이 성공적인지 확인합니다.
if response.status_code == 200:
    # 성공적인 경우, 응답 데이터를 파싱합니다.
    response_data = response.json()
    print(response_data)
else:
    # 실패한 경우, 오류 메시지를 출력합니다.
    print(f"Request failed. Status code: {response.status_code}")
```
