```python
import os
import requests
import json
from google.colab import files

# API 키 파일 업로드
uploaded = files.upload()

# 파일 이름을 입력하세요. 예를 들어, 'api_key.json'이라면 아래 코드를 'api_key.json'으로 변경하세요.
with open('open_ai_api_key.txt') as f:
    api_key = f.read().strip()

# API endpoint
url = "https://api.openai.com/v1/chat/completions"

# Headers
headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {api_key}"
}

# Body
data = {
    "model": "gpt-3.5-turbo",
    "messages": [{"role": "user", "content": "tell me about 5 delicious foods  "}],
    "temperature": 0.7
}

# Send POST request
response = requests.post(url, headers=headers, json=data)

# Get response JSON
response_json = response.json()

# Extract the first food description
content = response_json['choices'][0]['message']['content']
first_food_description = content.split("\n\n")[0]

# Print the first food description
print(first_food_description)
```
