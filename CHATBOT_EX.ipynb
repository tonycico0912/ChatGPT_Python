```python
import openai
from google.colab import files

# API 키 파일 업로드
uploaded = files.upload()

# 업로드된 파일명 가져오기
filename = list(uploaded.keys())[0]

# API 키 설정
with open(filename, 'r') as file:
    openai.api_key = file.read().strip()

# 질문 설정
q1 = 'what are the healthiest foods to eat everyday?'
q2 = 'make a recipe with those foods'
q3 = 'what is the amounts of kcals?'

# 질문 1에 대한 답변 생성
# 메시지 리스트를 생성합니다.
messages = [
    {
        'role': 'system',
        'content': 'Answer concisely as possible'
    },
    {
        'role': 'user',
        'content': q1
    }
]

# GPT-3.5-turbo 모델을 사용하여 대화 완성 생성
response = openai.ChatCompletion.create(
    model='gpt-3.5-turbo',
    messages=messages,
    temperature=0.6
)

# 생성된 응답에서 첫 번째 선택지의 메시지 내용을 가져옵니다.
bot_response_1 = response.choices[0].message.content

# 결과를 출력합니다.
print(bot_response_1)
print()
print('-' * 50)
print()

# 질문 2에 대한 답변 생성
# 이전 메시지들을 포함하는 메시지 리스트를 생성합니다.
messages = [
    {
        'role': 'system',
        'content': 'Answer concisely as possible'
    },
    {
        'role': 'user',
        'content': q1
    },
    {
        'role': 'assistant',
        'content': bot_response_1
    },
    {
        'role': 'user',
        'content': q2
    }
]

# GPT-3.5-turbo 모델을 사용하여 대화 완성 생성
response = openai.ChatCompletion.create(
    model='gpt-3.5-turbo',
    messages=messages,
    temperature=0.6
)

# 생성된 응답에서 첫 번째 선택지의 메시지 내용을 가져옵니다.
bot_response_2 = response.choices[0].message.content

# 결과를 출력합니다.
print(bot_response_2)
print()
print('-' * 50)
print()

# 질문 3에 대한 답변 생성
# 이전 메시지들을 포함하는메시지 리스트를 생성합니다.
messages = [
    {
        'role': 'system',
        'content': 'Answer concisely as possible'
    },
    {
        'role': 'user',
        'content': q1
    },
    {
        'role': 'assistant',
        'content': bot_response_1
    },
    {
        'role': 'user',
        'content': q2
    },
    {
        'role': 'assistant',
        'content': bot_response_2
    },
    {
        'role': 'user',
        'content': q3
    }
]

# GPT-3.5-turbo 모델을 사용하여 대화 완성 생성
response = openai.ChatCompletion.create(
    model='gpt-3.5-turbo',
    messages=messages,
    temperature=0.6
)

# 생성된 응답에서 첫 번째 선택지의 메시지 내용을 가져옵니다.
bot_response_3 = response.choices[0].message.content

print(bot_response_3)
print()
print('-' * 50)
print()
```
